```python
import pandas as pd
```

```python
train = pd.read_csv('WiC_dataset/train/train.data.txt', sep='\t', names=['target_word', 'pos', 'ind2ind', 'text_a',\
                                                                        'text_b'])
train_gold = pd.read_csv('WiC_dataset/train/train.gold.txt', sep='\t', names=['gold'])

train.head()
```

```python
dev = pd.read_csv('WiC_dataset/dev/dev.data.txt', sep='\t', names=['target_word', 'pos', 'ind2ind', 'text_a',\
                                                                        'text_b'])
dev_gold = pd.read_csv('WiC_dataset/dev/dev.gold.txt', sep='\t', names=['gold'])

dev.head()
```

```python
test = pd.read_csv('WiC_dataset/test/test.data.txt', sep='\t', names=['target_word', 'pos', 'ind2ind', 'text_a',\
                                                                        'text_b'])

test.head()
```

```python
train['label'] = train_gold['gold']
dev['label'] = dev_gold['gold']
```

```python
train.head()
```

```python
dev.head()
```

```python
train = train[['text_a', 'text_b', 'label']]
dev = dev[['text_a', 'text_b', 'label']]
test=test[['text_a', 'text_b']]
```

```python
train.label = train.label.map({'T': 1, 'F':0})
dev.label = dev.label.map({'T': 1, 'F':0})
```

```python
train.to_csv('train.tsv', sep='\t', header=False, index=False)
dev.to_csv('dev.tsv', sep='\t', header=False, index=False)
test.to_csv('test.tsv', sep='\t', header=False, index=False)
```
